```python
#qual è il prezzo medio per i prodotti di colore blu?

blue_filter = adw.loc[:,"Color"] == "Blue"
adw_blue = adw.loc[blue_filter]
mean_blue_price = adw_blue.loc[: , "DealerPrice"].mean()
mean_blue_price
```

```python
#qual è il prezzo medio per i prodotti di colore rosso o nero?

red_filter = adw.loc[:,"Color"] == "Red"
black_filter = adw.loc[:,"Color"] == "Black"
br_filter = red_filter | black_filter
adw_br = adw.loc[br_filter]
mean_br_price = adw_br.loc[:,"DealerPrice"].mean()
mean_br_price
```

```python
#qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?

size_flt = adw.loc[:,"Size"] == "42"
weight_flt = adw.loc[:,"Weight"] > 10
sw_flt = size_flt & weight_flt
adw_sw = adw.loc[sw_flt]
max_sw_price = adw_sw.loc[:,"DealerPrice"].max()
max_sw_price
```

```python
#qual è il nome inglese e il costo di produzione (StandardCost) di tutti i prodotti di taglia 42, peso oltre i 10 Kg e colore argento?

size_flt = adw.loc[:, "Size"] == "42"
weight_flt = adw.loc[:, "Weight"] > 10
silver_flt = adw.loc[:, "Color"] == "Silver"
ssw_flt = size_flt & weight_flt & silver_flt
adw_ssw = adw.loc[ssw_flt]
output = adw_ssw.loc[:,["EnglishProductName" , "StandardCost"]]

print(output)
```

```python
#Visualizziamo lo StandardCost e il DealerPrice degli ultimi 20 elementi del dataset: 
#quali sono le differenze? C'è un pattern? C'è qualche elemento che non lo segue?

adw_tail = adw.tail(20)
adw_tail_flt = adw_tail.loc[: , ["StandardCost","DealerPrice"]]
adw_tail_flt
```

```python
#quali sono le medie di charges rispetto a region? ci sono differenze significative?

#trasformo region in array

reg = np.array((ins["region"]).unique())
reg
```

```python
for i in reg:
    filtro = ins.loc[:, "region"] == i
    media = ins["charges"].loc[filtro].mean()
    print("Per la regione", i, "la media è", media)

#la parte ovest del paese paga di più in assicurazioni, mentre la zona più costosa è quella del sudest. Le differenze dei prezzi sono comunque trascurabili
```

```python
#rispetto a smoker e sex?

sex = np.array(ins["sex"].unique())

for i in sex:
    filtro = ins.loc[:, "sex"] == i
    media = ins["charges"].loc[filtro].mean()
    print("Il sesso", i, "paga in media", media)
    

```

```python

smoker = np.array(ins["smoker"].unique())

for i in smoker:
    filtro = ins.loc[:, "smoker"] == i
    media = ins["charges"].loc[filtro].mean()
    print("Il fumatore", i, "paga in media", media)
```

```python
#quali sono i pokemon leggendari?

import numpy as np

po_legend = po.loc[po["Legendary"]== True]

print(po_legend)
```

```python
#quali sono i pokemon leggendari di tipo 1 Grass?
po_grass = po_legend.loc[po["Type 1"] == "Grass"]
print(po_grass)
```

```python
#i leggendari di tipo Ice o Fire?

po_ice = po_legend.loc[po["Type 1"] == "Ice"]
po_fire = po_legend.loc[po["Type 1"] == "Fire"]
po_filter = po_ice.index.union(po_fire.index) #estraggo l'indice dei pokemon ice o fire e unisco i filtri
po_if = po.loc[po_filter]
print(po_if)
```
